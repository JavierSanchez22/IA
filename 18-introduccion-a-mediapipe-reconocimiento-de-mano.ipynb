```python
import cv2
import mediapipe as mp
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
```

```python
with mp_hands.Hands(
  static_image_mode=False,
  max_num_hands= 2,
    min_detection_confidence=0.5) as hands:
  image = cv2.imread('test.jpg')
  altura,ancho = image.shape
  image = cv2.flip(image, 1)
  image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  results = hands.process(image_rgb)
  print('Handedness:', results.multi_handedness)

  
```

```python
# Funcion para determinar si un dedo está levantando
def dedo_levantado(landmarks, dedo):
    if dedo == 'pulgar':
        return landmarks[mp_hands.HandLandmark.THUMB_TIP].x < landmarks[mp_hands.HandLandmark.THUMB_IP].x
    elif dedo == 'indice':
        return landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP].y < landmarks[mp_hands.HandLandmark.INDEX_FINGER_PIP].y
    elif dedo == 'medio':
        return landmarks[mp_hands.HandLandmark.MIDDLE_FINGER_TIP].y < landmarks[mp_hands.HandLandmark.MIDDLE_FINGER_PIP].y
    elif dedo == 'anular':
        return landmarks[mp_hands.HandLandmark.RING_FINGER_TIP].y < landmarks[mp_hands.HandLandmark.RING_FINGER_PIP].y
    elif dedo == 'meñique':
        return landmarks[mp_hands.HandLandmark.PINKY_TIP].y < landmarks[mp_hands.HandLandmark.PINKY_PIP].y
    return False
```

```python
# Funcion para determinar si un dedo esta levantado o no
def dedos_levantados(landmarks):
    dedos = ['pulgar', 'indice', 'medio', 'anular', 'meñique']
    return [dedo for dedo in dedos if dedo_levantado(landmarks, dedo)]
```

```python
if results.multi_hand_landmarks:
    for hand_landmarks in results.multi_hand_landmarks:
        mp_drawing.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
        print('Dedos levantados:', dedos_levantados(hand_landmarks.landmark))

        #Obtener las marcas como una lista
        landmarks = hand_landmarks.landmark

        #Definir los puntos de los dedos (segun mediapipe)
        pulgar_tipo = 4
        indice_tip = 8
        medio_tip = 12
        anular_tip = 16
        peque_tip = 20
        
```

```python
# Puntos para cada dedo (tip, dip, pip, mcp)
dedos = {
    'pulgar': (pulgar_tipo, 3, 2, 1),
    'indice': (indice_tip, 7, 6, 5),
    'medio': (medio_tip, 11, 10, 9),
    'anular': (anular_tip, 15, 14, 13),
    'meñique': (peque_tip, 19, 18, 17)
}
```

```python
#Verificar que dedos estan levantados
dedos_levantados = []
for nombre, (tip, dip, pip, mcp) in dedos.items():
    if dedo_levantado(landmarks, tip, dip, pip, mcp):
        dedos_levantados.append(nombre)
print('Dedos levantados:', dedos_levantados)
```
